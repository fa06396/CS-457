Function to compile all questions and remove null entries.

```python
def compileQuestions(start, end):
  temp = [i for i in range(start, end+1)]
  df2 = data.iloc[:,temp]
  df2.columns = df2.iloc[0].astype(str)
  df2 = df2[1:]
  colLst = list(df2.columns)
  # df2[colLst] = df2[colLst].astype(str)
  # print(colLst)
  # df3 = df2.loc[:, colLst].replace(1, pd.Series(df2.columns, df2.columns)).fillna('')
  df3 = df2.loc[:, colLst].replace("1", pd.Series(df2.columns, df2.columns)).fillna('')
  # df4 = combineCol(start, end, df3)
  return df3
```

```python
qDict = {
    1 : [12, 13],
    2 : [14, 17],
    3 : [18, 18],
    4 : [19, 19],
    5 : [20, 20],
    6 : [21, 21],
    7 : [22, 22],
    8 : [23, 26],
    9 : [27, 29],
    10 : [30, 34],
    11 : [35, 39],
    12 : [40, 44],
    13 : [45, 45]
}
```

```python
dFLst = []
for q in qDict:
  d = compileQuestions(qDict[q][0], qDict[q][1])
  dFLst.append(d)
```

```python
colLst = list(dFLst[0].columns)
dFLst[0] = dFLst[0].loc[:, colLst].replace(1, pd.Series(dFLst[0].columns, dFLst[0].columns)).fillna('')
```

```python
qNum = 1
while qNum != 14:
  colHeader = "Q" + str(qNum)
  dFLst[qNum-1].columns = [colHeader]
  qNum += 1
```
